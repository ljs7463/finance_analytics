# Using Lib

```python
stocks = yf.Ticker('AAPL')
```

```python
df = pd.DataFrame(stocks.history(start='2010-01-01', end='2022-09-23')['Close'])
df
```

```python
df.plot(figsize = (12, 5))
```

```python
plt.figure(figsize = (12, 5))
plt.fill_between(x= df.index, y1= df['Close'], alpha = 0.7)
plt.scatter(x =df.index, y = df['Close'], color = 'red', s =1)
plt.show()
```

# ìž¥,ë‹¨ê¸° ì¶”ì„¸(trend)

```python
fig, ax = plt.subplots(figsize = (14, 5))

# í•„ìš”ì—†ëŠ” ì„  ì œê±°
ax.spines[["left", "top", "right"]].set_visible(False)
# yì¶• ì„  ìƒì„±
ax.grid(axis="y", c="lightgray") # "axis = 'x'" ëŠ” xì¶• ì„¸ë¡œì„  gridìƒì„±

ax.plot(df.index, df['Close'])
ax.plot(df.index, df['Close'].rolling(5).mean(), color = 'red',label = '5ì¼ ì´ë™í‰ê· ì„ ',linewidth=2)
ax.plot(df.index, df['Close'].rolling(30).mean(), color = 'green',label = '30ì¼ ì´ë™í‰ê· ì„ ',linewidth=2)
ax.plot(df.index, df['Close'].rolling(90).mean(), color = 'yellow',label = '90ì¼ ì´ë™í‰ê· ì„ ',linewidth=2)
ax.plot(df.index, df['Close'].rolling(180).mean(), color = 'gray',label = '180ì¼ ì´ë™í‰ê· ì„ ',linewidth=2)
plt.legend()
plt.show()
```

# ì‹œê³„ì—´ ë¶„í•´ë¥¼ í†µí•œ íŒ¨í„´ íƒìƒ‰

ì‹œê³„ì—´ ë°ì´í„°ì—ëŠ” ì´ë ‡ê²Œ ì´ë™í‰ê· ìœ¼ë¡œ í™•ì¸í•  ìˆ˜ ìžˆëŠ” ì¶”ì„¸ë§Œ ìžˆëŠ” ê²ƒì´ ì•„ë‹ˆë¼ ê³„ì ˆì„±ê³¼ ìž”ì°¨ë¼ëŠ” ìš”ì†Œê°€ ìžˆìŠµë‹ˆë‹¤. ê³„ì ˆì„±ì€ ê´€ì¸¡ì¹˜(observed, ì£¼ê°€)ì—ì„œ ì¶”ì„¸(trend)ë¥¼ ì œê±°í•˜ê³  ë‚¨ì€ ë°ì´í„° ì¤‘ ë°˜ë³µë˜ëŠ” íŒ¨í„´ì„ ìž¡ì•„ì„œ ë³´ì—¬ì£¼ë©° ìž”ì°¨(residual, resid)ëŠ” ê³„ì ˆì„±ê¹Œì§€ ì œê±°ë˜ê³  ë‚¨ì€ ê°’ì„ ë³´ì—¬ì¤ë‹ˆë‹¤. ë”°ë¼ì„œ ìž”ì°¨ì—ì„œëŠ” íŠ¹ì • ë°˜ë³µ íŒ¨í„´, ì»¤ì§€ê±°ë‚˜ ìž‘ì•„ì§€ëŠ” ì›€ì§ìž„, ê¸‰ê²©ížˆ íŠ€ëŠ” êµ¬ê°„ ë“±ì´ ë‚¨ì•„ ìžˆìœ¼ë©´ ì•ˆ ë©ë‹ˆë‹¤. ëª¨ë“  íŒ¨í„´ê³¼ ì›€ì§ìž„ì´ ì¶”ì¶œë˜ê³  ë‚¨ì€ 'ë‚˜ë¨¸ì§€'ëŠ” random, ì¦‰ ë¬´ìž‘ìœ„ì  ë³€ë™ì„±ì´ì–´ì•¼ í•˜ê¸° ë•Œë¬¸ìž…ë‹ˆë‹¤. â€‹

```python

from statsmodels.tsa.seasonal import seasonal_decompose 
res = seasonal_decompose(df ,model = 'addictive',period=30)


plt.figure(figsize=(12,6))

plt.subplot(411)
plt.xticks([])
plt.ylabel('observed')
plt.plot(res.observed)

plt.subplot(412)
plt.xticks([])
plt.ylabel('trend')
plt.plot(res.trend)

plt.subplot(413)
plt.xticks([])
plt.ylabel('seasonal')
plt.plot(res.seasonal)

plt.subplot(414)
plt.ylabel('resid')
plt.plot(res.resid)

# ì „ì²´ ê·¸ëž˜í”„ì— ëŒ€í•œ íƒ€ì´í‹€ì€ suptitleë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
plt.suptitle('seasonal decomposed')
# í¼ì ¸ìžˆëŠ” ê·¸ëž˜í”„ê°„ ê°„ê²©ì„ íƒ€ì´íŠ¸í•˜ê²Œ ìž¡ì•„ì¤ë‹ˆë‹¤.
plt.tight_layout()
plt.show()
```

- ì¶”ì„¸ëŠ” 20ë…„ì´í›„ í¬ê²Œ ìƒìŠ¹í•œê²ƒì„ ì•Œ ìˆ˜ ìžˆìœ¼ë©° 2020ë…„ ì´í›„ ë¬´ìž‘ìœ„ì„±ì´ ë†’ì•„ì§€ê³  ìžˆìŒì„ ë³´ì—¬ì£¼ê³  ìžˆë‹¤, ì¦‰ ì–´ë– í•œ ì‚¬ê±´ì´ ìžˆìŒì„ ì§ìž‘í•´ ë³¼ ìˆ˜ ìžˆë‹¤.

```python
df_2022 = df.reset_index()
df_2022 = df_2022[df_2022['Date'].dt.year==2022]
df_2022 = df_2022.set_index('Date')
```

```python
# 2022ë…„ ì´í›„ ë°ì´í„°ë§Œ ë¶„í•´í•´ë´…ë‹ˆë‹¤.
res = seasonal_decompose(df_2022['Close'], period=30)

plt.figure(figsize=(12,6))

plt.subplot(411)
plt.xticks([])
plt.ylabel('observed')
plt.plot(res.observed)

plt.subplot(412)
plt.xticks([])
plt.ylabel('trend')
plt.plot(res.trend)

plt.subplot(413)
plt.xticks([])
plt.ylabel('seasonal')
plt.plot(res.seasonal)

plt.subplot(414)
plt.ylabel('resid')
plt.plot(res.resid)

# ì „ì²´ ê·¸ëž˜í”„ì— ëŒ€í•œ íƒ€ì´í‹€ì€ suptitleë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
plt.suptitle('seasonal decomposed')
# í¼ì ¸ìžˆëŠ” ê·¸ëž˜í”„ê°„ ê°„ê²©ì„ íƒ€ì´íŠ¸í•˜ê²Œ ìž¡ì•„ì¤ë‹ˆë‹¤.
plt.tight_layout()
plt.show()
```
